## Missing values

```python
data.isna().sum()
```

```python
data.loc[data['workclass'].str.contains('\?')]
```

```python
data['workclass'].str.contains('\?').sum()
```

```python
category_columns = [] 
```

```python
data.columns
```

```python
for col in data.columns: 
    if data[col].dtype != 'int64': 
        category_columns.append(col)
```

```python
category_columns
```

```python
for col in category_columns: 
    have_interrogation = data[col].str.contains('\?').any()
    if have_interrogation: 
        print(f'Na coluna {col} há interrogações')
```

```python
for col in category_columns: 
    data[col] = data[col].replace(' ?', np.nan)
```

```python
data.isna().sum()
```

```python
data.fillna(np.nan, inplace = True) 
data.iloc[0:30]
```

```python
data['workclass'].fillna(data['workclass'].mode().iloc[0], inplace = True) 
data['occupation'].fillna(data['occupation'].mode().iloc[0],  inplace = True) 
data['native-country'].fillna(data['native-country'].mode().iloc[0], inplace = True)
```

```python
data.isna().sum()
```

### renomeando em loop

```python
upper_columns = []

for col in data.columns: 
    upper_columns.append(str(col).upper().strip()) 


data.columns = upper_columns 
data.columns
```

```python
data.columns 
data.reindex(labels = ['WORKCLASS','AGE','FINAL-WEIGHT', 'EDUCATION', 'EDUCATION-NUM',
       'MARITAL-STATUS', 'OCCUPATION', 'RELATIONSHIP', 'RACE', 'SEX',
       'CAPITAL-GAIN', 'CAPITAL-LOOS', 'HOUR-PER-WEEK', 'NATIVE-COUNTRY',
       'INCOME'] , axis = 1)
```

## Criando novas colunas 

```python
data['HOUR-PER-MONTH'] = data['HOUR-PER-WEEK'] * 4  
data['HOUR-PER-MONTH'] 


data.columns 

data = data.reindex(labels = ['AGE', 'WORKCLASS', 'FINAL-WEIGHT', 'EDUCATION', 'EDUCATION-NUM',
       'MARITAL-STATUS', 'OCCUPATION', 'RELATIONSHIP', 'RACE', 'SEX',
       'CAPITAL-GAIN', 'CAPITAL-LOOS', 'HOUR-PER-WEEK','HOUR-PER-MONTH', 'NATIVE-COUNTRY',
       'INCOME'] , axis = 1) 

data.head()
```

```python
data['HIGH-EDUCATION-LEVEL'] = data['EDUCATION-NUM'] > 11 
data.head()
```

```python
data['INCOME'] = data['INCOME'].str.strip()
```

```python
dicionario_map = {
  '<=50K': 'Low', 
  '>50K': 'High'
} ## criamos as condições e definiremos as colunas baseadas nestas condições 


data['CATEGORY-INCOME'] = data['INCOME'].map(dicionario_map) 
```

```python
data.head()
```

```python
data['SEX'] = data['SEX'].str.strip()
```

```python
sex_code = {
    'Male': 1, 
    'Female': 2
} 

data['SEX-CODE'] = data['SEX'].map(sex_code)
```

```python
data.head()
```

### Tratamento com funções lambda

```python
data['CAPITAL-GAIN-DOLLARS'] = data['CAPITAL-GAIN'].map(lambda x: f'USD {x}')
```
